```python
import gym 
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(rme)
print(rme.observation_space)
print(rme.action_space)
print(rme.maze)
print(rme.maze.shape)
print(rme.hole_loc)
print(rme.goal_loc)
print(rme.wall_loc)
print(rme.hole_reward)
print(rme.goal_reward)
print(rme.step_reward)
print(rme.prob)
```
